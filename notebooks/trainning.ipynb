```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt")
```

```python
results = model.train(
    data="/content/dataset/data.yaml",
    epochs=120,
    patience=20,           # early stopping
    imgsz=640,
    batch=-1,              # tự động chọn batch size
    device=0,
    workers=2,
    cache="disk",          # an toàn hơn cache RAM
    verbose=True,
    project="/content/drive/MyDrive/yolo_runs",
    name="chicken_duck_train",
    save=True,
    plots=True,
    save_period=10,

    # Optimizer
    optimizer="AdamW",
    lr0=0.001,
    lrf=0.01,
    weight_decay=0.0005,
    warmup_epochs=3,

    # Augmentation
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    flipud=0.3,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.1,
)
```
